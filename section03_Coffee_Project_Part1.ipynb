# Coffee Consumption Vs. Prices

Compare coffee prices with coffee consumption in the following countries.

Is there a relationship between coffee consumption and the price these countries pay?

Plot a dual axis bar chart of price vs consumption totals for these countries.

```python
coffe_imports_upload = files.upload()
```

```python
consumption = pd.read_csv("imports.csv" ,
                          index_col=[0]
                          )
consumption
```

```python
agg_consumption = consumption.mean(axis = 1)
agg_consumption.name = "Avg_Imports"
agg_consumption.head()
```

```python
retail_prices_upload = files.upload()
```

```python
prices = pd.read_csv("retail-prices.csv" , index_col=[0])
prices.head()
```

```python
agg_prices = prices.mean(axis = 1)
agg_prices.name = "Avg_Price"
agg_prices.head()
```

```python
prices_consumption = pd.concat([agg_prices , agg_consumption] , axis = 1)
prices_consumption.dropna(inplace=True)
prices_consumption.sort_values(by = ["Avg_Imports" , "Avg_Price"] , ascending=[False ,False])
```

```python
fig , ax = plt.subplots(figsize = (16,10) , ncols=1 , nrows=2)

width = 0.35
x = np.arange(len(prices_consumption.index))

ax[0].bar(x - width/2,
       prices_consumption["Avg_Price"],
       width = width,
       label = "Average Price",
       color ="blue"
       )
ax[0].set_title("Paid Prices vs Consumptions of the selected nations" , fontsize = 16)
ax[0].set_ylabel("Average Price Paid Per KG - USD" , color = "blue")
ax[0].tick_params(axis = 'y' , color = "blue" , labelcolor = "blue")
ax[0].grid(True,)

ax01 = ax[0].twinx()
ax01.bar(x + width/2 ,
       prices_consumption["Avg_Imports"],
       width = width,
       label = "Average Imports",
       color ="orange"
       )
ax01.set_xticks(x , labels = prices_consumption.index)
ax01.set_ylabel("Consumption" , color = "orange")
ax01.tick_params(axis = 'y' , color = "orange" , labelcolor = "orange" )
fig.legend()
ax01.grid(False)

ax[1].scatter(x =prices_consumption["Avg_Price"] , y = prices_consumption["Avg_Imports"])
ax[1].set_title("Avg_Import vs Avg_Imports")
ax[1].set_xlabel("Average Price")
ax[1].set_ylabel("Average Imports")

plt.show()
```
